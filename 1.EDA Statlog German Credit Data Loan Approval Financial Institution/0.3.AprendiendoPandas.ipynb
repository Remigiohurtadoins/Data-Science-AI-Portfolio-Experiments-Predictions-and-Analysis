## <span style="color:BLUE">2. MANEJO DE DATASET Y VISUALIZACIÓN</span>

```python
#Información del dataset
#url: https://www.kaggle.com/ronitf/heart-disease-uci
#url: https://archive.ics.uci.edu/ml/datasets/heart+Disease
#Esta base de datos contiene 76 atributos, pero todos los experimentos publicados se refieren al uso de un subconjunto de 14
#de ellos. El campo "objetivo" se refiere a la presencia de enfermedad cardíaca en el paciente. 

#age: age in years
#sex: (1 = male; 0 = female)
#cp: chest pain type
#trestbps: resting blood pressure (in mm Hg on admission to the hospital)
#chol: serum cholestoral in mg/dl
#fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
#restecg: resting electrocardiographic results
#thalach: maximum heart rate achieved
#exang: exercise induced angina (1 = yes; 0 = no)
#oldpeak: ST depression induced by exercise relative to rest
#slope: the slope of the peak exercise ST segment
#ca: number of major vessels (0-3) colored by flourosopy
#thal: 3 = normal; 6 = fixed defect; 7 = reversable defect
#target: 1 or 0

import pandas as pd
dataset = "Datasets/heart.csv"
#la primera fila del dataset contiene los nombres de las columnas (variables). Suelen haber datasets que no contienen esta fila

#Cargar el dataset en un DataFrame de Pandas desde un archivo
df = pd.read_csv(dataset)

#Para renombrar columnas
df.columns = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','target']

df.head()#Muestra las primeras 5 filas de datos. Recordar que en Python el índice va desde 0
#df.tail()#Muestra las últimas 5 filas de datos.

#Para conocer tipo de datos de cada variable
#print (df.dtypes)
#Para cambiar el tipo de datos de las variables indicadas
#df[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','target']] = df[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','target']].apply(pd.to_numeric)
```

```python
features=['age','sex','cp','trestbps','chol']
# Seleccionar algunas variables en un ndarray de numpy. 
# Esto permitirá utilizar las funciones de NumPy
x = df.loc[:, features].values
# Seleccionar variable objetivo "target" en un ndarray de numpy
y = df.loc[:,['target']].values
print(x)
type(x)
#print(y)
```

## <span style="color:BLUE">3. MANEJO DE DATASET, DATAFRAME Y VISUALIZACIÓN AVANZADA</span>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline 
dirResultados="Datasets/Resultados.csv"
dataResultados=pd.read_csv(dirResultados,sep=';')
dataResultados.head()
#print (dataResultados.dtypes)
```

```python
valoresK=[20,100,150,200]
Nmin=10
Nmax=15
dataResultados=dataResultados.loc[(dataResultados['N'] >= Nmin) & (dataResultados['N'] <= Nmax)]#Filtrar por algunos valores de N
dataResultados=dataResultados.loc[(dataResultados['K'].isin(valoresK))]#Filtrar por algunos valores en K

#Escribir dataframe en un archivo csv
dataResultados.to_csv("Datasets/ResultadosFiltrados.csv", sep=";",index = False) #sep es el separado, por defector es ","

dataResultados.head()
```

```python
ResultadosPorTecnica=dataResultados.groupby(['METHOD', 'K'])
ResultadosPorTecnica=ResultadosPorTecnica.mean()
print(ResultadosPorTecnica)
ResultadosPorTecnica=ResultadosPorTecnica.groupby('METHOD')
```

```python
for name, group in ResultadosPorTecnica:
    print(name)
    print(group)
    print("\n")
```

```python
for name, group in ResultadosPorTecnica:
    print(name)
    valores = group.PRECISION
    
    #SIN EFECTO SUAVIZADO
    if name=='PC': plt.plot(valoresK, valores, marker='x', linestyle='--', color='C1', ms=5, label = name) 
    if name=='RUV': plt.plot(valoresK, valores, marker='s', linestyle='-', color='C2', ms=5, label = name)    
    if name=='RAP': plt.plot(valoresK, valores, marker='o', linestyle='--', color='C9', ms=3, label = name)
    if name=='RAF': plt.plot(valoresK, valores, marker='.', linestyle='-', color='C0', ms=10, label = name)
    
    plt.title("Resultados")
    plt.xlabel("Número de Grupos (K)")   # Inserta el título del eje X 
    plt.ylabel("Precision")   # Inserta el título del eje Y
    plt.legend(loc="upper right")
```

```python
for name, group in ResultadosPorTecnica:
    print(name)
    valores = group.PRECISION
  
    #EFECTO SUAVIZADO: GENERAR PUNTOS CON UNA REGRESIÓN ENTRE LOS VALORES QUE TENEMOS
    numeroPuntos=20  #representa el número de puntos para generar entre T.min and T.max 
    xnew = np.linspace(min(valoresK),max(valoresK),numeroPuntos) #linspace genera puntos en un rango
    spl = make_interp_spline(valoresK, valores, k=3) #BSpline object 
    power_smooth = spl(xnew) 
    
    if name=='PC': plt.plot(xnew, power_smooth, marker='x', linestyle='--', color='C1', ms=5, label = name) 
    if name=='RUV': plt.plot(xnew, power_smooth, marker='s', linestyle='-', color='C2', ms=5, label = name)    
    if name=='RAP': plt.plot(xnew, power_smooth, marker='o', linestyle='--', color='C9', ms=3, label = name)
    if name=='RAF': plt.plot(xnew, power_smooth, marker='.', linestyle='-', color='C0', ms=10, label = name) 
    
    plt.title("Resultados")
    plt.xlabel("Número de Grupos (K)")   # Inserta el título del eje X 
    plt.ylabel("Precision")   # Inserta el título del eje Y
    plt.legend(loc="upper right")
```
